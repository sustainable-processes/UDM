# Data formatting for plotting graphs in overleaf

```python
import pandas as pd
```

```python
performance = pd.read_csv('performance.csv')
```

```python
performance
```

```python
#random_forest
for i in range(len(performance)):
    if performance['regressor'][i]=='random_forest':
        print('(', performance['real'][i], ',', performance['predicted'][i], ')')
```

```python
#linear regression
for i in range(len(performance)):
    if performance['regressor'][i]=='linear_regression':
        print('(',performance['real'][i],',',performance['predicted'][i],')')
```

```python
#gaussian process regressor
for i in range(len(performance)):
    if performance['regressor'][i]=='gaussian_process':
        print('(', performance['real'][i], ',', performance['predicted'][i], ')')
```
